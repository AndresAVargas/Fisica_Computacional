# Problema 2: Caminatas Aleatorias

La caminata del "borracho" consiste modelar la trayectoria de una persona si esta da pasos en direcciones aleatorias conforme transcurre el tiempo. 

La forma mas sencilla de generar este tipo de trayectorias no es dificil; suponiendo que la caminata tiene un principio y un final, basta seguir los siguientes pasos.

- Seleccionar un punto $x(t=0)$ donde el caminante A comenzara su partida, asi como el tiempo que tomara todo el movimiento $t$. 

- Si cada paso es de la misma longitud y solo se puede dar un paso por cada t, identifiquemos con el numero 1 a un paso a la derecha, mientras que -1 es para un paso a la izquierda.

- La posicion del borracho $x(t)$ al tiempo t estara determinada por su posicion en el instante anterior mas la eleccion aleatoria de ir a la derecha o a la izquierda.
$$ x(t) =  x(t-1) + \begin{Bmatrix}
1, -1
\end{Bmatrix}.$$

### 1 
Lleva las instrucciones anteriores a codigo.  Realiza tres caminatas aleatorias, cada una representa a un borracho que parte siempre del mismo lugar y  termina su caminanta con la misma cantidad de pasos (1000), visualiza cada una de ellas en una sola grafica.

```python
x=0 # Posicion Borracho 1
y=0 # Posicion Borracho 2
z=0 # Posicion Borracho 3
direccion=[1,-1]   #direccion derecha o izquierda la definimos como 1 y -1 y los agregamos a una lista
caminata1=[]     #lista vacia para guardar los pasos del borracho 1
caminata2=[]     #lista vacia para guardar los pasos del borracho 2
caminata3=[]     #lista vacia para guardar los pasos del borracho 3
for t in range(1000):   #empieza nuestro for mientras t este entre 1 y 1000 hara lo de abajo
    paso1 = direccion[randint(0,2)]    #paso del borracho 1 el cual se tomara aleatoreamente de la lista direccion 
    paso2 = direccion[randint(0,2)]    #paso del borracho 2 el cual se tomara aleatoreamente de la lista direccion 
    paso3 = direccion[randint(0,2)]    #paso del borracho 3 el cual se tomara aleatoreamente de la lista direccion 
    x+=paso1                      #sumamos el elemto al azar a la posición del borracho 1
    y+=paso2                      #sumamos el elemto al azar a la posición del borracho 2
    z+=paso3                      #sumamos el elemto al azar a la posición del borracho 3
    caminata1.append(x)           #agregamos x a la lista caminata del borracho 1
    caminata2.append(y)           #agregamos y a la lista caminata del borracho 2
    caminata3.append(z)           #agregamos z a la lista caminata del borracho 3
print(caminata1)                  #imprimimos la lista de caminata del borracho 1 
print(caminata2)                  #imprimimos la lista de caminata del borracho 2
print(caminata3)                  #imprimimos la lista de caminata del borracho 3
plot(caminata1, color='b', label='Borracho 1')     #graficamos la caminata de nuestro borracho 1 le damos color y nombre a su grafica
plot(caminata2, color='r', label='Borracho 2')    #graficamos la caminata de nuestro borracho 2 le damos color y nombre a su grafica
plot(caminata3, color='purple', label='Borracho 3')  #graficamos la caminata de nuestro borracho 3 le damos color y nombre a su grafica
title('Caminata aleatoria de 3 borrachos')  #Le ponemos titulo a la grafica
ylabel('Cantidad de pasos dados a la derecha o a la izquierda') #le ponemos titulo al eje y
xlabel('Numero de pasos')   #le ponemos titulo al eje x
legend()    #mostramos las etiquetas de label
```

(1) En papel, convierte 0.1 a binario, demuestra que el error por redondeo es aproximadamente de $\frac{3}{80}$. ¿Cuál será el error de redondeo si sumamos 0.1+0.1+0.1 en binario?

```python
#(2)
n=int(input("Cuantos decimales quieres:"))   #pedimos los decimales que quiere el usuario
def binario(x):  #definimos la funcion binario
    c=[]       #definimos la lista vacia
    y = int(x)   #tomamos la parte entera de x
    while y != 1:  #mientras la parte entera de x sea distinto de 1
        d = y % 2   #se define d igual al residuo de el entero de x entre 2
        c.append(d)  #lo guardamos en una lista c
        y = y//2    #dividimos enteramente el entero de x en 2
    else:
        c.append(y)   #agregamos y en una lista 
    c.reverse()   #"volteamos la lista"
    z=x-y        #restamos la parte entera a x
    a=[]   #creamos otra lista vacia
    for i in range(n):    #comienza un for, mientras i este en el entero que nos dio el usuario entonces
        z*=2     #la parte fraccionaria lo multiplicamos por 2
        b=int(z) % 2   #definimos el residuo del entero de z entre 2
        a.append(b)   #lo guardamos en la lista a
    return [c,a]   #regresa la funcion la lista c y a 
print(binario(70.42)) #imprimimos y llamamos la funcion
```

```python
#(2)
def decimal(x,y):   #definimos la funcion decimal
    x1=0      #definimos un contador para nuestra parte entera
    y1=0      #definimos otro contador para la parte fraccionaria
    x.reverse()   #volteamos la lista que se le da a la funcion
    y.reverse()   #volteamos la lista que se le da a la funcion
    for i in range(len(x)): #mientras i este en el rango de la longitud de la lista x
        x1 += int((x[i]))*(2**i)   #a el contador de sumamos el entero del i-esimo entero multiplicado por dos elevado a la potencia de i
    for i in range(len(y)):     #mientras i este en el rango de la longitud de la lista y
        y1 = ( y[i] +y1 ) /2     #al contador le sumamos el i-esimo elemento mas el contador entre 2
    return x1+y1   #devolvemos la suma de los contadores x1 es nuestra parte entera y y1 la parte fraccionaria
print(decimal([1, 0, 0, 0, 1, 1, 0], [0, 1, 1, 0, 1, 0, 1, 1, 1, 0]))  #imprimimos y llamamos a la funcion
```

```python
#def Error porcentual
def error_porcentual(teo,real): #definimos la funcion, pedimos 2 variables para que funcione
    return (abs((teo-real)/teo)) *100    #regresamos el valor absoluto de la resta de las dos variables entre la variable teorica
print(error_porcentual(9.81,9.90)) #imprimimos y llamamos a la funcion para un ejemplo del error porcentual
```

\begin{equation}

	x_{1,2}=\frac{2c}{-b\mp\sqrt{b^2-4ac}}

```python
#Definimos Búsqueda incremental
def incremental(f,a,b,dx):
    x0=a;f0=f(x0)
    x1=x0+dx;f1=f(x1)
    contador=0.0
    while f0*f1>0.0:
        if x1>=b: return ('nada') #salgo del bucle infinito
        x0=x1;f0=f1
        x1=x0+dx;f1=f(x1)
        contador=contador+1
    return [x1,contador]
```

```python
#Definimos el Metodo de bisección
def biseccion(f,a,b,epsilon=1e-6):
    contador=0
    dx=abs(b-a)
        #return 'Da un intervalo correcto'
    while dx>epsilon:
        x=(a+b)/2.0
        if (f(a)*f(x))<0:
            b=x #izquierda
        elif (f(x)*f(b))<0:
        #else:
            a=x #derecho
        else:
            dx=epsilon/2
            print('El intervalo no tiene raíz')
        dx=abs(b-a) #renombrar intervalo
        contador=contador+1
    return [x, contador]
```

```python
#Definimos el Método de Newton-Raphson
def newton(f,df,x,epsilon=1e-6):
    dx=2*epsilon
    contador = 0
    while dx> epsilon and contador<1000:
        x1=x-f(x)/df(x)
        dx=abs(x-x1)
        x=x1
        contador=contador+1
    return [contador,x,dx]
```

```python
#Definimos el Método del Secante
def secante(f, x0, x1, delta=1e-9):
    f0 = f(x0)
    f1 = f(x1)
    res = [[x0, f(x0), np.nan], #nan es el paso inicial del Data Frame
           [x1, f(x1), x1 - x0],]
    while True: 
        x2 = x1 - f1 * (x1 - x0) / (f1 - f0) #secante
        f2 = f(x2)
        res.append([x2, f2, x2-x1])
        if abs(f2) < delta:
            break
        f0 = f1
        f1 = f2
        x0 = x1
        x1 = x2
    return res
```

```python
#Definimos el Método de Ridders
from math import *
#no uso el modulo error (no lo necesito)
def ridders(f,a,b,tol=1e-6):
    contador=1
    fa=f(a)
    if fa==0: return [a,contador]
    fb=f(b)
    if fb==0: return [b,contador]
    if fa*fb>0.0: return('No es el intervalo con la raíz incluida') # o print() o error.error()
    for i in range(30): 
        c=0.5*(a+b);fc=f(c) #tomamos el punto medio, sería la x3
        s=sqrt(fc**2-fa*fb)
    if s==0.0: return None
    dx=(c-a)*fc/s
    if(fa-fb)<0.0:dx=-dx #selecciono los signos
    x=c+dx; fx=f(x)
    #aquí acaba el metodo
    #pruebo la convergencia
    x_vieja=x
    if i>0:
        if abs(x-x_vieja) < tol*max(abs(x),1.0): return [x,contador]
    #x_vieja=x
    if fc*fx>0.0:
        if fa*fx<0.0: b=x;fb=fx
        else: a=x;fa=fx
    else:
        a=c;b=x;fa=fc;fb=fx
    contador=contador+1
    return None
    print('Muchas iteraciones')
```
