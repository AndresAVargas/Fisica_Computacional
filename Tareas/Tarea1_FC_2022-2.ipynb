## La probabilidad de los dados

En una feria local, las personas juegan los dados contra el propietario.  El juego consiste en lanzar dos dados y si la suma de ellos es igual a 7 el propietario paga \\$8.00 al jugador,  en caso de obtener un numero diferente el jugador paga \\$2.00 al propietario.



### 1
Tu trabajo consiste en simular a una persona que jugara los dados 100 veces consecutivas. Con un bucle, simula todos los juegos y cuenta por cada tirada si se gana o se pierde.  Estima cuanto dinero se gano y cuanto se perdio. Tambien establece la probabilidad de ganar en el juego de los dados al jugar 100 veces, da una conclusion con tus propias palabras sobre el juego tomando en cuenta todos los resultados.


```python
#Programando el Juego
dado_1=[1,2,3,4,5,6] #Definimos el dado 1 
dado_2=[1,2,3,4,5,6] #Definimos el dado 2
g=0    #El dinero ganado, lo definimos con la variable g
p=0    #El dinero perdido, lo definimos con la variable p
P=0    #Contaremos las veces que se gano para calcular la probabilidad frecuencial , lo definiremos con la variable P
for lanzamientos in range(100):     #Se realizaran 100 lanzamientos     
    lanzamiento1 = dado_1[randint(0,6)]    #encontramos el i-esimo elemento i=aleatorio entre el 1 y el 6
    lanzamiento2 = dado_2[randint(0,6)]    #encontramos el i-esimo elemento i=aleatorio entre el 1 y el 6
    if lanzamiento1 + lanzamiento2 == 7:    #condicional if, si la suma del dado 1 y el 2 son igual a siente entonces ocurre lo de abajo:
        print("Lanzamiento",lanzamientos+1,":","Ganaste $8.00!")  #Al cumplirse el if imprime el numero de lanzamiento y un mensaje
        g+=8            #Al cumplirse el if suma 8 a nuestra variable g, el cual es el dinero ganado
        P+=1            #Con esto contaremos las veces que se gano
    else:              #Si no se cumple el if, entra la condicional else
        print("Lanzamiento",lanzamientos+1,":","Perdiste, paga $2.00!") #Al entrar el else imprime un mensaje   
        p+=2      #Ademas al entrar el else suma 2 a nuestra variable p, el cual es el dinero perdido
print("En total ganaste: $", g)  #Al final del bucle, es decir, cuando se hacen los 100 lanzamientos imprime cuanto dinero se gano
print("Y en total has perdido: $", p) #Ademas se imprime cuando dinero se perdio
print("Se ganó un total de:", P )  #Imprimimos las veces que se ganó
print("La probabilidad de ganar es de", P, "%, en decimal el de:", P/100) #Imprimimos la probabilidad frecuencial de ganar en porcentaje (como se lanzo 100 veces la probabilidad es igual a las veces que se ganó) y en forma de decimal
```

# Problema 2: Caminatas Aleatorias

La caminata del "borracho" consiste modelar la trayectoria de una persona si esta da pasos en direcciones aleatorias conforme transcurre el tiempo. 

La forma mas sencilla de generar este tipo de trayectorias no es dificil; suponiendo que la caminata tiene un principio y un final, basta seguir los siguientes pasos.

- Seleccionar un punto $x(t=0)$ donde el caminante A comenzara su partida, asi como el tiempo que tomara todo el movimiento $t$. 

- Si cada paso es de la misma longitud y solo se puede dar un paso por cada t, identifiquemos con el numero 1 a un paso a la derecha, mientras que -1 es para un paso a la izquierda.

- La posicion del borracho $x(t)$ al tiempo t estara determinada por su posicion en el instante anterior mas la eleccion aleatoria de ir a la derecha o a la izquierda.
$$ x(t) =  x(t-1) + \begin{Bmatrix}
1, -1
\end{Bmatrix}.$$

### 1 
Lleva las instrucciones anteriores a codigo.  Realiza tres caminatas aleatorias, cada una representa a un borracho que parte siempre del mismo lugar y  termina su caminanta con la misma cantidad de pasos (1000), visualiza cada una de ellas en una sola grafica.

```python
x=0 # Posicion Borracho 1
y=0 # Posicion Borracho 2
z=0 # Posicion Borracho 3
direccion=[1,-1]
caminata1=[]
caminata2=[]
caminata3=[]
for t in range(1000):
    paso1 = direccion[randint(0,2)]
    paso2 = direccion[randint(0,2)]
    paso3 = direccion[randint(0,2)]
    x+=paso1
    y+=paso2
    z+=paso3
    caminata1.append(x)
    caminata2.append(y)
    caminata3.append(z)
print(caminata1)
print(caminata2)
print(caminata3)
plot(caminata1, color='b', label='Borracho 1')
plot(caminata2, color='r', label='Borracho 2')
plot(caminata3, color='purple', label='Borracho 3')
title('Caminata aleatoria de 3 borrachos')
ylabel('Cantidad de pasos dados a la derecha o a la izquierda')
xlabel('Numero de pasos')
legend()
```
