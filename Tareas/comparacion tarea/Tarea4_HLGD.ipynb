##4.- Digamos que tenemos la siguiente ecuación:

$$
\frac{dx}{dt} = \frac{1}{x^2+t^2}
$$

Y la queremos resolver, $t$ va de cero a infinito y ponemos como condición inicial en $x$ que sea igual a uno cuando $t=0$ Encuentra $g(x,u)$ y escribe su expresion en la casilla de abajo.

Del desarrollo realizado en el notebook 2 relacionado a ecuaciones diferenciales ordinarias recordemos que se tiene la expresión: $\frac{dx}{dt}=f(x,t)$ y que por regla de la cadena $\frac{dx}{dt}=\frac{dx}{du}\frac{du}{dt}=f(x,t) \Rightarrow \frac{dx}{du}\frac{du}{dt}=f(x,t) \Rightarrow \frac{dx}{du}=\frac{dt}{du}f(x,t)$ donde bajo un cambio de variable se consideró $u=\frac{t}{1+t}$ y $t=\frac{u}{1-u} \Rightarrow \frac{dt}{du}=\frac{1}{(1-u)^{2}}$ que sustituyendo en la derivada respecto a u y sustituyendo también el cambio de variable de t a u en la función f:

$$\frac{dx}{du}=\frac{1}{(1-u)^{2}}f(x,\frac{u}{1-u})$$

En este caso paarticular se tiene que $f(x,t)=\frac{1}{x^2+t^2}$ que bajo el cambio de variable $\Rightarrow f(x,\frac{u}{1-u})=\frac{1}{x^{2}+(\frac{u}{1-u})^{2}}=\frac{1}{\frac{x^{2}(1-u)^{2}+u^{2}}{(1-u)^{2}}}=\frac{(1-u)^{2}}{(x(1-u))^{2}+u^{2}}=\frac{(1-u)^{2}}{(x-xu)^{2}+u^{2}}$ y sustituyendo $f(x,\frac{u}{1-u})$ en $\frac{dx}{du} \Rightarrow$ 

$$\frac{dx}{du}=\frac{1}{(1-u)^{2}}\frac{(1-u)^{2}}{(x-xu)^{2}+u^{2}}$$
$$\Rightarrow \frac{dx}{du}=\frac{1}{(x-xu)^{2}+u^{2}}$$

$$\therefore g(x,u)=\frac{1}{(x-xu)^{2}+u^{2}}$$

##5. Ahora acompleta el código (Runge-Kutta de orden 4) para resolver la ecuación anterior.

```python
from pylab import *

def g(x,u):
    return (1)/(((x-x*u)**2)+((u)**2))
    #######llenar#######

a,b=0,1 ####poner rango####

N=100
h=(b-a)/N
lista_u=arange(a,b,h)
lista_t,lista_x=[],[]
x=1.0 #####aqui valor inicial#####
for u in lista_u:
    lista_t.append(u/(1-u)) #para plotear con respecto a t
    lista_x.append(x)
    k1=h*g(x,u)
    k2=h*g(x+0.5*k1,u+0.5*h)
    k3=h*g(x+0.5*k2,u+0.5*h)
    k4=h*g(x+k3,u+h)
    x+=(k1+2*k2+2*k3+k4)/float(6)
plot(lista_t,lista_x, color="red")
title('Runge-Kutta de orden 4 con cambio de variable cuando t tiende a infinito')
xlim(0,100)
grid(True)
xlabel('t')
ylabel('x(t)')
show()
```

##7.- Módelo depredador y presa (Lotka-Volterra)

Es interesante para representar interacciones de depredador-presa en sistemas biológicos.

Tradicionalmente se habla de conejos (presa, $x$) y zorros (depredador, $y$). Las cantidades $x,y$ se piensan normalmente en los miles.

Los conejos se reproducen en una tasa proporcional a su población misma, pero son devorados en una tasa proporcional a su población y a la población de zorros:


$$
\frac{dx}{dt} = \alpha x - \beta x y
$$

Donde $\alpha$ y $\beta$ son constantes. A su vez los zorros se reproducen a una tasa proporcional a la tasa que devoran conejos, pero se mueren a una tasa proporcional a su población misma (por edad avanzada).

$$
\frac{dy}{dt} = \gamma x y - \delta y
$$

a) Escribe un programa para resolver estas ecuaciones usando Runge-Kutta de cuarto orden para el caso $\alpha=1$, $\beta=\gamma=0.5$ y $\delta = 2$ empezando de la condición inicial $x=y=2$ (i.e. 2000 conejos y dos mil zorros). Has que el programa haga una gráfica mostrando tanto $x$ como $y$ como función del tiempo usando como mismo eje $t$ que vaya desde cero hasta 30. NOTA: A pesar de que no hay dependencia explícita de $t$ es conveniente escribir la función de la forma $f(\mathbf{r},t)$ incluyendo la variable temporal para no tener que reescribir nada.

b) plotea zorros vs conejos.

c) describe lo que sucede usando ambas gráficas en términos de zorros y conejos.

```python

def f(r,t):
   x=r[0]
   y=r[1]
   fx=alpha*x-beta*x*y
   fy=gamma*x*y-delta*y
   return array([fx,fy],float)

a,b=0.0,30.0
N=1000
h=(b-a)/N
lista_t=arange(a,b,h)
lista_x,lista_y=[],[]
r=array([2.0,2.0],float) #condiciones iniciales
alpha,beta,gamma,delta=1.0,0.5,0.5,0.2
for t in lista_t:
    lista_x.append(r[0])
    lista_y.append(r[1])
    k1=h*f(r,t)
    k2=h*f(r+0.5*k1,t+0.5*h)
    k3=h*f(r+0.5*k2,t+0.5*h)
    k4=h*f(r+k3,t+h)
    r+=(k1+2*k2+2*k3+k4)/float(6)
plot(lista_t,lista_x,label='conejos')
plot(lista_t,lista_y,label='zorros', color="brown")
title('Modelo depredador presa. Población de zorros y conejos en función del tiempo.')
xlabel('t')
ylabel('zorros/conejos (Contabilizados en miles)')
legend()
show()
plot(lista_x,lista_y, color="red") #espacio fase
title('Espacio-fase conejos y zorros. Contabilizados por miles en ambos ejes')
xlabel('conejos')
ylabel('zorros')
show()
```

De ambas gráficas se confirma un gran entrelazamiento entre las poblaciones de conejos y zorros, lo anterior se visualiza del hecho de que en principio ambas especies tienen la misma población y se puede apreciar que la población de conejos comienza disminuir de manera drástica, dichas muertes de conejos en este sistema se pueden atribuir a la caza de la cual son sujetos por sus depredadores los zorros, ahora bien, se observa que conforme disminuye drásticamente la población de conejos de igual manera drásticamente aumenta la población de zorros, pero sería ilógico pensar que por cada conejo muerto espontáneamente surja un zorro, lo que podría suceder es que debido a que los zorros se alimentan de los conejos se encuentran en mejores condiciones físicas para sobrevivir ante las inclemencias del clima o bien ante ataques a su vez de otras especies, prolongando su esperanza de vida y aumentando las posibilidades de poder recorrer grandes distancias y encontrar alguna pareja con la cual aparearse y tener crías. 

Entonces, si bien los conejos son el eslabón más débil en comparación con los zorros se observa que los zorros dependen en gran medida de los conejos, debido a que cuando ha disminuido la población de conejos de manera muy considerable se vuelve menos probable que la totalidad de los zorros puedan alimentarse satisfactoriamente, por lo que un gran número de éstos (de los zorros) morirán de hambre o serán mucho más vulnerables a eventualidades de su ecosistema lo cual de igual manera ocasionará la muerte de una buena parte de su población, ya que podría ser el caso incluso que entre más escaso sea el alimento los mismos lobos tengan que pelear entre sí para consumir lo poco que hay aún cuando se encuentren en manadas. 

Dicho lo anterior, una vez que la población de lobos comienza a disminuir por obvias razones habrá menor cantidad de depredadores que puedan acechar a los conejos y debido a la alta tasa de reproducción en los conejos éstos volverán aumentar su población, y cuando su población aumenta, los pocos zorros presentes tienen un gran abasto de alimento, permitiendo que el ciclo se repita.

De las ecuaciones diferenciales más conocidas:

$$
\frac{dx}{dt}=\sigma(y-x), \frac{dy}{dt}=rx-y-xz, \frac{dz}{dt}=xy-bz
$$

Donde $\sigma, r$ y $b$ son constantes.

Estudiadas por Edward Lorenz en 1963, para estudiar patrones de clima simplificados. Son famosas por ser de los primeros ejemplos de caos determinista, apariencia de movimiento aleatorio a pesar de que no hay aleatoridad en las ecuaciones.

```python
def f(r,t):
   x=r[0]
   y=r[1]
   z=r[2]
   fx=sigma*(y-x)
   fy=(m*x)-y-(x*z)
   fz=(x*y)-(n*z)
   return array([fx,fy,fz],float)

a,b=0,50
N=10000
h=(b-a)/N
lista_t=arange(a,b,h)
lista_x,lista_y,lista_z=[],[],[]
r=array([0.0,1.0,0.0],float) #condiciones iniciales
sigma,m,n=10,28,(8/3)
for t in lista_t:
    lista_x.append(r[0])
    lista_y.append(r[1])
    lista_z.append(r[2])
    k1=h*f(r,t)
    k2=h*f(r+0.5*k1,t+0.5*h)
    k3=h*f(r+0.5*k2,t+0.5*h)
    k4=h*f(r+k3,t+h)
    r+=(k1+2*k2+2*k3+k4)/float(6)
#plot(lista_t,lista_x)
plot(lista_t,lista_y, color="blue")
#plot(lista_t,lista_z)
title('$y$ en función del tiempo')
xlabel('t')
ylabel('$y(t)$')

show()
plot(lista_x,lista_z, color="red")
title('Atractor de Lorenz.')
xlabel('$x(t)$')
ylabel('$z(t)$')
show()
```

## 9.-Filtro Low-Pass

Las ecuaciones son

$$V_{in}-V_{out}=IR, Q=CV_{out}, I=\frac{dQ}{dt}$$

De forma equivalente tenemos entonces:

$$\frac{dV_{out}}{dt} = \frac{1}{RC} (V_{in}-V_{out})$$

$$V_{in} =\left\{
\begin{array}{ll}
      1  & \mathrm{si}\, floor(2t)\, \mathrm{es\, par}\\
      -1 & \\
\end{array} \right.$$

```python
from pylab import *
RC=0.01
def V_in(t):
    if int(2*t)%2==0:
        return 1
    else:
        return -1

def f(x,t):
    return (V_in(t)-x)/float(RC)

a,b=0.0,10.0
N=1000
h=(b-a)/N
x=0.0
tpoints=arange(a,b,h)
xpoints=[]

for t in tpoints:
    xpoints.append(x)
    k1=h*f(x,t)
    k2=h*f(x+0.5*k1,t+0.5*h)
    k3=h*f(x+0.5*k2,t+0.5*h)
    k4=h*f(x+k3,t+h)
    x+=(k1+2*k2+2*k3+k4)/float(6)
plot(tpoints,xpoints,color="blue")
title('Filtro low-pass.')
xlabel('t')
ylabel('x(t)')
show()
```

##10.- Escribir expresión para un objeto que se suelta una altura $h$. Completar el código para plotear también la expresión analítica.

Para realizar lo requerido recordemos que la segunda Ley de Newton corresponde a $\bar{F}=m\bar{a}$ donde tenemos la aceleración como una consecuencia de una fuerza ejercida sobre el objeto, en este caso se puede considerar que la única fuerza que se ejerce sobre el objeto de masa $m$ que cae (despreciando resistencia al viento y las velocidades terminales que esto produce considerando que el experimento no se realiza en el vacío) es la fuerza gravitacional dada por $\bar{F}=-mg\hat{z}$ y recordando que la aceleración se define como la razón de cambio de la velocidad respecto al tiempo $\Rightarrow \bar{a}=\frac{d\bar{v}}{dt}$ que sustituyendo: $-mg\hat{z}=m\frac{d\bar{v}}{dt}$ siendo una ecuación diferencial separable $\Rightarrow -g\hat{z}dt=d\bar{v}$ que integrando a ambos lados $\Rightarrow \int_{0}^{t}-g\hat{z}dt=\int_{v_{0}}^{v} d\bar{v} \Rightarrow \bar{v}=v_{0}-gt\hat{z}$ donde $t_{0}$ es constante de integración, a su vez la velocidad se define como la razón de cambio de la posición respecto al tiempo $\Rightarrow \frac{d\bar{z}}{dt}=v_{0}-gt\hat{z}$ que también es una ecuación separable $$\Rightarrow d\bar{z}=(v_{0}-gt\hat{z})dt$$
$$\Rightarrow \int_{h_{0}}^{z} d\bar{z}=\int_{0}^{t}(v_{0}-gt\hat{z})dt$$
$$\bar{z}=(-g\frac{t^{2}}{2}+v_{0}t+h_{0})\hat{z}$$
Donde $h_{0}$ corresponde a la altura inicial y $v_{0}$ a la velocidad inicial 

```python
from pylab import *

def f(r,t):
    g=-9.81
    x=r[0]
    v=r[1]
    fx=v
    fv=g
    return array([fx,fv],float)

a = 0.0
b = 10.0
N = 1000
h = (b-a)/N

lista_v=[]
lista_x = []
lista_t = []
t=0
r=([30.,0.]) #cond inicial
while r[0]>0.0:
    k1=h*f(r,t)
    k2=h*f(r+0.5*k1,t+0.5*h)
    k3=h*f(r+0.5*k2,t+0.5*h)
    k4=h*f(r+k3,t+h)
    r+=(k1+2*k2+2*k3+k4)/6.
    t+=h
    lista_x.append(r[0])
    lista_v.append(r[1])
    lista_t.append(t)
plot(lista_t,lista_x,color="blue")
title('Caída libre de un objeto por método numérico.')
xlabel('tiempo')
ylabel('altura')
show()
analitica=[]
def caidalibre_analitica(h0,t):
  return -9.81*((t**2)/2)+h0 
for i in lista_t:
  analitica.append(caidalibre_analitica(30,i))

plot(lista_t,analitica,color="red")
title('Caída libre de un objeto graficado con la expresión analítica.')
xlabel('tiempo')
ylabel('Posición al caer.')
  
```

##14.- Implementa una función que use el método leapfrog para el caso de m=k=1 con fuerza externa  F=0 , pocición inicial  x0=0,v0=1  variando  γ=0,0.01,0.1 . Y grafica la posición y energía en función del tiempo.

```python
N = 10000
t = linspace(0,300,N)
dt = t[1] - t[0]


def leapfrog(F,x0,v0,gamma):
    x = zeros(N)
    v = zeros(N)
    E = zeros(N)    


    x[0] = x0
    v[0] = v0


    fac1 = 1.0 - 0.5*gamma*dt
    fac2 = 1.0/(1.0 + 0.5*gamma*dt)

    for i in arange(N-1):
        #Aqui va tu codigo de leapfrog
        v[i + 1] = fac1*fac2*v[i] - fac2*dt*x[i] + fac2*dt*F[i]
        x[i + 1] = x[i] + dt*v[i + 1]
        E[i] += 0.5*(x[i]**2 + ((v[i] + v[i+1])/2.0)**2)

    E[-1] = 0.5*(x[-1]**2 + v[-1]**2)

    return x,v,E


F = zeros(N)
x1,v1,E1 = leapfrog(F,0.0,1.0,0.0) # x0 = 0.0, v0 = 1.0, gamma = 0.0
x2,v2,E2 = leapfrog(F,0.0,1.0,0.01) # x0 = 0.0, v0 = 1.0, gamma = 0.01
x3,v3,E3 = leapfrog(F,0.0,1.0,0.1) # x0 = 0.0, v0 = 1.0, gamma = 0.5

figure()
subplot(211)
plot(t,x1)
plot(t,x2)
plot(t,x3)
ylabel("x(t)")

subplot(212)
plot(t,E1,label=r"$\gamma = 0$")
plot(t,E2,label=r"$\gamma = 0.01$")
plot(t,E3,label=r"$\gamma = 0.1$")
ylabel("E(t)")

xlabel("Time")
rcParams['legend.fontsize'] = 14.0
legend(loc="upper left")

show()
```

```python
#Graficando variando nuévamente el parámetro \gamma
F = zeros(N)
x1,v1,E1 = leapfrog(F,0.5,0.5,0.0) # x0 = 0.0, v0 = 1.0, gamma = 0.0
x2,v2,E2 = leapfrog(F,0.5,0.5,1.5) # x0 = 0.0, v0 = 1.0, gamma = 0.01
x3,v3,E3 = leapfrog(F,0.5,0.5,0.5) # x0 = 0.0, v0 = 1.0, gamma = 0.5

figure()
subplot(211)
plot(t,x1)
plot(t,x2)
plot(t,x3)
ylabel("x(t)")

subplot(212)
plot(t,E1,label=r"$\gamma = 0$")
plot(t,E2,label=r"$\gamma = 1.5$")
plot(t,E3,label=r"$\gamma = 0.5$")
ylabel("E(t)")

xlabel("Time")
rcParams['legend.fontsize'] = 14.0
legend(loc="upper left")

show()
```
