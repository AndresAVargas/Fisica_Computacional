# Visualizaci√≥n de vectores

```python
#suma de vectores
from numpy import *
from pylab import *

a=array([-1.7,1.5])
b=array([2.1,1.9])

c=[]
for i in range(len(a)):
    x=a[i]+b[i]
    c.append(x)
c=array(c)
print(c)


quiver([-1.7], [1.5],  color=['r'], scale_units='x', scale=1) #vector a
quiver([2.1], [1.9],  color=['b'], angles='xy', scale_units='xy', scale=1) #vector b
quiver(c[0],c[1],color=['g'], angles='xy', scale_units='xy', scale=1) #vector c

xlim(-4, 4)
ylim(-4, 4)
grid()

show()
```

### Funci√≥n gradiente

$$
z=xe^{-x^2-y^2}
$$

```python
from pylab import *
x = arange(-2,2.2,0.2)
y = arange(-2,2.2,0.2)

X, Y = meshgrid(x, y)
z = X*exp(-X**2 -Y**2)
dx, dy = gradient(z)

quiver(X,Y,dx,dy)
show()
```

Por √∫ltimo, se har√° un ejemplo en el que se tendr√° 4 puntos que
representan v√≥rtices del campo vectorial, es decir, cuatro puntos centrales en los que los vectores parecen girar alrededor de √©stos. La funci√≥n que describe los cuatro v√≥tices de este ejemplo es:

$$
\vec{F}=\sin x \cos y \hat{i}- \cos x \sin y \hat{j}
$$

```python
from pylab import *

#Definimos los colores en funcion del gradiente
x = np.arange(-2,2.2,0.2)
y = np.arange(-2,2.2,0.2)
X, Y = np.meshgrid(x, y)
z = X*exp(-X**2 -Y**2)
dx, dy = gradient(z)
n = -5
color_arreglo = sqrt(((dx-n)/2)**2 + ((dy-n)/2)**2)


#Hacemos la visualizaci√≥n de los vectores
x = arange(0,2*pi+2*pi/20,2*pi/20)
y = arange(0,2*pi+2*pi/20,2*pi/20)

X,Y = meshgrid(x,y)

u = sin(X)*cos(Y)
v = -cos(X)*sin(Y)

quiver(X,Y,u,v,color_arreglo)

show()
```

### Histogramas 

Para graficar valores discretos o distribuciones estad√≠sticas es √∫til usar un histograma. Tambi√©n son √∫tiles para graficar una distribuci√≥n de n√∫meros que cruzan un rango de posibles valores. Ejemplo: Tomar una lista de n√∫meros aleatorios y ver cuantas veces de ocurrencia que presentan. Los n√∫meros aleatorios ser√°n generados con  distribuciones gaussianas.

```python
from random import *
numeros=normal(size=1000)
```

```python
import matplotlib.pyplot as p
from random import *
numeros = normal(size=1000)
p.hist(numeros)
p.title("Histograma Gaussiano")
p.xlabel("Valor")
p.ylabel("Frecuencia")
p.show()
```

1. Nos referimos a "bins"  como el n√∫mero de distintas categor√≠as. El default en python es 10 bins. 
2. Si queremos cambiar la resoluci√≥n de nuestra gr√°fica, es decir, cambiar el n√∫mero de bins a 20 usamos p.hist(numeros, bins=20)
3. Si queremos graficar una distribuci√≥n de probabilidad en lugar de un conteo de frecuencias, se tendr√° que dividir entre una constante de normalizaci√≥n. Por lo general, se quiere que el 100% de probabilidad se encuentre en 1. p.hist(numeros, bins=20,normed=True) 
4. Si queremos cambiar la apariencia del histograma, que los cuadros no est√©n llenados, sino que s√≥lo se vean los escalones usamos la opci√≥n histtype p.hist(numeros, bins=20,normed=True,histtype='step'

### Ejemplo
Hacer un histograma con la distribuci√≥n gaussiana y una distribuci√≥n uniforme de n√∫meros aleatorios desde -3 a 3 en rojo con 50% de transparencia sobre el color az√∫l de la distribuci√≥n gaussiana: 

```python
from pylab import *
from numpy.random import normal, uniform
numeros = normal(size=1000)
uniformes = uniform(low=-3, high=3, size=1000)
hist(numeros, bins=20, histtype='stepfilled', color='b',label='Gaussiana')
hist(uniformes, bins=20, histtype='stepfilled', color='r', alpha=0.5, label='Uniforme')
title("Histograma Gaussiana/Uniforme")
xlabel("Valor")
ylabel("Probabilidad")
legend()
show()
```

### Graficas en 3-d

Para habilitar la graficaci√≥n en 3-d en Matplotlib se tiene que habilitar un kit de herramientas extra *toolkit* extra llamado mplot3d. El kit de herramientas mplot3d agrega capacidades simples de trazado en 3D a matplotlib al proporcionar un objeto de ejes que puede crear una proyecci√≥n 2D de una escena 3D. El gr√°fico resultante tendr√° el mismo aspecto que los gr√°ficos 2D normales. El kit de herramientas mplot3d incluye formatos interactivos que brindan la capacidad de rotar y hacer zoom en la escena 3D. Se puede rotar la escena 3D simplemente haciendo clic y arrastrando la escena. El zoom se realiza haciendo clic con el bot√≥n derecho en la escena y arrastrando el mouse hacia arriba y hacia abajo. Tenga en cuenta que no se usa el bot√≥n de zoom como se usar√≠a para gr√°ficos 2D normales.

```python
%matplotlib notebook
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

mpl.rcParams['legend.fontsize'] = 14 #

fig = plt.figure()
ax = fig.gca(projection='3d')

theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)

r = z**2 + 1
#r=1
x = r * np.sin(theta)
y = r * np.cos(theta)

ax.plot(x, y, z, label='helicoidal')
ax.legend()
```

```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np


fig = plt.figure()
ax = plt.axes(projection="3d")

def z_function(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = z_function(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")

ax.plot_wireframe(X, Y, Z, color='green') 
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z,cmap='twilight') #winter, hot ,twilight, summer, spring, magma, Spectral
#rainbow,prism, plasma
#ax.set_title('surface');
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# Generar el mallado del toro #torus (tori)
angulo = np.linspace(0, 2 * np.pi, 32)
theta, phi = np.meshgrid(angulo, angulo)
#Generamos el toro de revoluci√≥n
r, R = .25, 1.
X = (R + r * np.cos(phi)) * np.cos(theta)
Y = (R + r * np.cos(phi)) * np.sin(theta)
Z = r * np.sin(phi)

fig = plt.figure()
ax = fig.gca(projection = '3d')
ax.plot_surface(X, Y, Z, color = 'r')


u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = 2 * np.outer(np.cos(u), np.sin(v))
y = 2 * np.outer(np.sin(u), np.sin(v))
z = 2 * np.outer(np.ones(np.size(u)), np.cos(v))

ax.plot_wireframe(x, y, z)

plt.show()
```

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm

fig = plt.figure()
ax = fig.gca(projection='3d')
X, Y, Z = axes3d.get_test_data(0.05)
ax.plot_surface(X, Y, Z, rstride=8, cstride=8, alpha=0.3)
cset = ax.contour(X, Y, Z, zdir='z', offset=-100, cmap=cm.coolwarm)
cset = ax.contour(X, Y, Z, zdir='x', offset=-40, cmap=cm.coolwarm)
cset = ax.contour(X, Y, Z, zdir='y', offset=40, cmap=cm.coolwarm)

ax.set_xlabel('X')
ax.set_xlim(-40, 40)
ax.set_ylabel('Y')
ax.set_ylim(-40, 40)
ax.set_zlabel('Z')
ax.set_zlim(-100, 100)

plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import colors

fig = plt.figure()
ax = fig.gca(projection='3d')

N = 100
y = np.ones((N,1))
x = np.arange(1,N + 1)
z = 5*np.sin(x/5.)

cn = colors.Normalize(min(z), max(z)) # creates a Normalize object for these z values
for i in range(N-1):
    ax.plot(x[i:i+2], y[i:i+2], z[i:i+2], color=plt.cm.jet(cn(z[i])))

plt.show()
```
