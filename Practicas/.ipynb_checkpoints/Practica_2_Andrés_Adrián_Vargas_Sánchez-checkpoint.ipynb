## 1.- 
La velocidad de un cohete que se dispara verticalmente desde la superficie de la tierra puede aproximarse por la siguiente formula:

$$ v = -gt + u \ln \frac{M}{M-kt}.$$

donde 

$u = 2510 m/s $ es la velocidad de escape relativa al cohete.

$M = 2.8x10^{5}kg$ La masa del cohete.

$k = 13.3x10^{3} kg/s$ La proporcion del consumo de combustible.


#### a)
Determina el tiempo cuando el cohete logra la velocidad del sonido $v=355m/s$

```python
#a)
v=355     #Definimos las variables, la velocidad del sonido
g=9.81    #la aceleracion de la gravedad 
u=2510    #la velocidad de escape
M=2.8e5   #la masa del cohete
k=13.3e3    #la proporcion del consumo de combustible

def f(t):
    x=v-(-g*t+u*log(M/(M-k*t)))          #Definimos una funcion donde regresara el valor de x que es el tiempo 
    return x

t=arange(0.1,500)                           #hacemos una secuencia de numeros del tiempo que ingresaremos en la funcion que definimos
axhline(0, color="r", linestyle="--")     #agregamos una linea punteada de color rojo en el eje x sobre el 0
title('Velocidad vs Tiempo')         #Agregamos el titulo a la grafica
ylabel('Velocidad [m/s]')     #agregamos el titulo del eje y
xlabel('Tiempo [s]')     #agregamos el titulo del eje x
plot(t,f(t))          #graficamos
 
def secante(f, x0, x1, delta):     #definimos la funcion secante para encontrar la raiz
    start=time.time()
    f0 = f(x0)
    f1 = f(x1)
    contador = 0
    res = [[x0, f(x0), np.nan], #nan es el paso inicial del Data Frame
           [x1, f(x1), x1 - x0],]
    while True: 
        x2 = x1 - f1 * (x1 - x0) / (f1 - f0) #secante
        f2 = f(x2)
        res.append([x2, f2, x2-x1])
        if abs(x2-x1)<delta:
            break
        f0 = f1
        f1 = f2
        x0 = x1
        x1 = x2
        contador=contador+1
    end=time.time()
    tiempo=end-start
#     error=abs(raiz-x2)
    return res, [x2,contador, tiempo]
T=secante(f, 0.1, 3, 1e-9)[-1][0]     # guardamos el elemento 0 de la ultima lista que nos arroja la funcion secante la cual es la raiz que buscamos
print("El tiempo cuando el cohete logra la velocidad del sonido es de", T, "s") #imprimimos el resultado 
```

## 2.-
En el campo de la construccion, una viga voladiza es una estructura que esta soportada solo uno de sus lados. La ecuacion que describe la frecuencia de oscilacion de la viga es la siguiente:

$$ f(\beta) = \cosh(\beta)\cos(\beta) + 1. $$

donde 

$\beta_{n}^{4} = (2\pi f_{n})^{2} \frac{mL^{3}}{EI}$

$f_{n}$ La frecuencia enesima

$m, L$ La masa y longitud de la viga

$E$ El modulo de elasticidad

$I$ El momento de inercia de la seccion transversal

#### a) 
Determina dos frecuencias para la viga tales que $f(\beta)=0$.  Considera L= 0.9m,  con una seccion trnasversal rectangular de 25mm de ancho y 25 mm de alto.   La densidad de masa para el acero es de 7850 kg/m**3 y su modulo de elasticidad de 200 GPa.


```python
#a)
def f(b):               #definimos nuestra funcion que describe la frecuencia de oscilacion de la viga
    y=cosh(b)*cos(b)+ 1  
    return y
b=arange(0,5,0.01)   #definimos la secuencia de numeros que tomara el valor de beta
plot(b,f(b))      #graficamos la funcion contra beta
plot(b,zeros(len(b)),"--r")   #graficamos una linea en el cero del eje x 
title("Frecuencia vs Beta") #agreganos titulo a la grafica 
ylabel('Frecuencia [Hz]')   #agregamos el titulo del eje y
xlabel('Beta')   #agregamos el titulo del eje x

def secante(f, x0, x1, delta):    #definimos otro metodo para calcular raiz
    start=time.time()
    f0 = f(x0)
    f1 = f(x1)
    contador = 0
    res = [[x0, f(x0), np.nan], #nan es el paso inicial del Data Frame
           [x1, f(x1), x1 - x0],]
    while True: 
        x2 = x1 - f1 * (x1 - x0) / (f1 - f0) #secante
        f2 = f(x2)
        res.append([x2, f2, x2-x1])
        if abs(x2-x1)<delta:
            break
        f0 = f1
        f1 = f2
        x0 = x1
        x1 = x2
        contador=contador+1
    end=time.time()
    tiempo=end-start
#     error=abs(raiz-x2)
    return res, [x2,contador, tiempo]

B1=secante(f,1.5,2.4,10e-5)[-1][0]       #guardamos las raices encontrada en las variables B1 y B2 son el elemento 0 de la ultima lista
B2=secante(f,4,5,10e-5)[-1][0]
d=7850      #definimos las variables, densidad de masa
L=0.9   #longitud de la viga
m=d*L*0.025*0.025   #la masa de la viga
I=(0.025**4)/12   #momento de inercia
E=2e11     #modulo elastico
def frecuencia(B):   #definimos la función que nos calculara nuestras frecuencias 
    f=(1/(2*pi))*sqrt(((B**4)*E*I)/(m*(L**3)))
    return f
print("Las frecuencias donde son cero son:") #imprimimos los resultados
print("Frecuencia 1:", frecuencia(B1))
print("Frecuencia 2:", frecuencia(B2))
```

## 3.-
La siguiente figura  muestra el ciclo termodinamico de un motor.  La eficiencia del motor esta dada por la siguiente expresion:

$\eta = \frac{\ln(T_{2}/T_{1}) + (1 - T_{2}/T_{1})} {\ln(T_{2}/T_{1}) - (1 - T_{2}/T_{1})/(\gamma -1) }.$

#### a) 
Encuentra el valor de $T_{1}$ que logra una eficiencia del 30%, considera $T_{2}$ = 400K, $\gamma = 5/3$ Que tipo de gas hace funcionar al motor?

```python
#a)
T2=400   #definimos T2
G=5/3    #definimos gamma
def f(T1):    #definimos la funcion que nos dara el 30 porciento de la eficiencia menos la funcion de la eficiencia
    x= (log(T2/T1) + (1- (T2/T1))) / (log(T2/T1)-(1-(T2/T1))/(G-1))
    return 0.3- x
T1=arange(1,10000)       #definimos una secuencia de numeros que tomara la temperatura 1 
plot(T1,f(T1))   #graficamos
title('Eficiencia vs Temperatura T1') #le damos titulo a la gráfica
ylabel('Eficiencia')  #le damos titulo al eje y
xlabel('Temperatura [K]') #le damos titulo al eje X
plot(T1,zeros(len(T1)),"--r")   #graficamos una linea en el eje x sobre el 0

#Definimos el Método de bisección
def biseccion(f,a,b,eps=1e-6):
    contador=0 #Definimos el contador de pasos
    dx=abs(b-a) #Distancia del intervalo

    while dx>eps:
        x=(a+b)/2.0 #Dividimos el intervalo
        if (f(a)*f(x))<0: #Si cambia d signo en la primera mitad
            b=x #Redefinimos el intervalo
        elif (f(x)*f(b))<0: #Si cambia de signo en el lado derecho
            a=x #Redefinimos el intervalo
        else:
            dx=eps/2
            print('El intervalo no tiene raíz, hay más de una o no cruza el eje')
            break
        dx=abs(b-a) #Renombramos el intervalo
        contador=contador+1
    return [x, contador]

print("La temperatura T1 que logra una eficiencia de 30% es: T1= ",biseccion(f, 2000,4000,1e-6)[0],"K") #imprimos el resultado
```

##### R=
El gas que hace funcionar a este motor es un gas monoatómico, podemos notarlo en su valor de $\gamma$

## 4.-

Un problema elemental en mecanica cuantica, es encontrar las energias de estados ligados en un pozo de potencial con profundidad P. Las energias de los estados ligados son soluciones  de las siguientes ecuaciones trascendentales:

$\sqrt{P - E_{B}}\tan\left ( \sqrt{P - E_{B}} \right ) = \sqrt{E_{B}}.$ Para funciones de onda pares

$\frac{\sqrt{P - E_{B}}}{\tan\left ( \sqrt{P - E_{B}} \right )} = \sqrt{E_{B}}.$ Para funciones de onda impares

#### a)
Encuentra las soluciones en ambos casos para 2 valores de la profundidad del pozo. 


```python
# a)
def f_par(E_b):  #definimos la funcion que nos da la raiz de E_b para ondas pares menos la raiz de E_b
    return sqrt(P - E_b)*tan(sqrt(P-E_b))- sqrt(E_b)

def f_impar(E_b): #definimos la funcion que nos da la raiz de E_b para ondas impares menos la raiz de E_b
    return sqrt(P - E_b)/(tan(sqrt(P-E_b)))- sqrt(E_b)
P=13 #damos el valor de P

E_par=arange(0,3,0.01)   #definimos una secuencia de valores que tomara E_b en la funcion par
E_impar=arange(4,13,0.01) #definimos una secuencia de valores que tomara E_b en la funcion impar

plot(E_par, f_par(E_par))  #graficamos la funcion par
axhline(0, color="r", linestyle="--")  #graficamos una linea en el 0 sobre el eje x
show()
plot(E_impar, f_impar(E_impar)) #graficamos la funcion impar
axhline(0, color="r", linestyle="--") #graficamos una linea en el 0 sobre el eje x
show()

print("Para P=", P," el valor de E_b es:") #imprimos los resultados
print("Para la funcion par: ", biseccion(f_par,1,2,1e-6)[0] )
print("Para la funcion impar no hay valor para E_b")
```
