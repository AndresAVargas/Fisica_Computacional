### 1) 
Un empleado trabaja 40 horas a la semana a razon de 3 mil pesos por semana, si la paga por hora aumenta en 1.5 veces por cada hora extra trabajada en la semana, elabora un programa  que calcule el salario mensual del trabajador, si conoce las horas que ha trabajado en cada semana.

```python
x = float(input("Horas de trabajo?"))  #}Preguntamos al usuario cuantas horas trabajo el empleado y el valor que de es convertido en flotante
sh=3000/40                             #Damos valor a la variable sh que calcula cuanto gana en 1 hr
if x<=40:                              #Condicional que va a checar si el valor dado es menor o igual a 40
    S=x*sh                             #Si el condicional se cumple multiplicara el valor dado por sh dando el valor S
    print("Tu sueldo es de", S)        #Luego imprimira un mensaje con la variable S que es el salario ganado en 1 semana
else:                                  # Condicional contraia al if, si no se cumple que x es menor igual a 40 entonces se cumple esta condicional
    S=sh*40 + sh*(1.5)*(x-40)          #Al cumplirse esta condicional se calculara el salario base con sh* 40y se le sumaran las horas extras que trabaje con sh*1.5*(x-40), x-40 le resta las horas base para saber cuantas horas extra trabajo
    print("Tu sueldo es de", S)        #Luego imprime un mensaje con el valor calculado anteriorimente de la variable S
```

Elabora un programa que genere todos los años bisiestos entre los años 1900 y 3000,  considerando que 1900 fue bisiesto.  Puedes hacerlo con los dos bucles for y while

El problema de la liebre y la tortuga.

En un lejano bosque se dispusieron a competir en una carrera de 400 metros una liebre y una tortuga. La carrera tuvo momentos importantes en diferentes intervalos de tiempo, donde ambos competidores mostraron diferentes estrategias, las cuales se describen a continuación:

- Entre $[0\,min, 2\,min)$ la velocidad de la liebremfue de  $V_l= 0.9 \,m/s$, mientras que la tortuga comenzo la carrera a $V_t= 0.3 \,m/s$.


- Entre $[2\,min, 7\,min)$ la liebre se relajo un poco, reduciendo su velocidad a $V_l= 0.2 \,m/s$, mientras que la de la tortuga aumento un poco  a $V_t= 0.6 \,m/s$. En este intervalo, la liebre tuvo cuidado, cada vez que la tortuga estaba a menos de 2 metros, su velocidad aumentaba a $V_l= 1.2 \,m/s$.  La tortuga siempre mantuvo el paso.


- Pasados 7 minutos, los corredores guardaron energia por 5 minutos,  $[7\,min, 12\,min)$ reduciendo sus velocidades, la tortuga corrio a $V_t= 0.4 \,m/s$, mientras que la liebre prefierio ponerse a dormir.


- En los minutos finales de la carrera $[12\,min, 15\,min)$, la liebre aumento el paso al verse en peligro por perder la competencia, corriendo a $V_l= 1.65 \,m/s$,  la tortuga tambien dio su ultimo esfuerzo, alcanzando su velocidad maxima de $V_t= 0.6 \,m/s$.


- La carrera puede terminar antes de los 15 min, cuando alguno de los dos animales haya pasado los 400 metros. 


Como nadie puede creer los resultado de la carrera, elabora un programa para dar validez al resultado, imprimiendo el ganador, el tiempo que duro la carrera, y las distancias recorridas por ambos competidores. Considera los intervalos de tiempo que se proporcionaron. 

$\textbf{Sugerencia:}$ Utiliza un bucle/ciclo que tome diferentes valores de tiempo, por cada vez que se repita el bucle/ciclo el tiempo habra avanzado un segundo. En cuanto uno de los dos competidores supere los 400 metros interrumpe el ciclo. 

```python
a,b,c,d =2*60,7*60,12*60,15*60   #Intervalos de tiempo 
t,l=0,0                          #Distancia de la tortuga y liebre
k=0                              #Tiempo en segundos
while k<d:                       #Declaramos el bucle que continuará mientras la codicion k menor a d sea verdadera
    if k<a:                      #Usamos un condicional if con la condicion de que k sea menor a a
        l+=0.9                   #Al cumplirse la condicion sumamos a la variable l 0.9
        t+=0.3                   #Al cumplirse la condicion sumamos a la variable t 0.3
    elif k>=a and k<b:           #Al ya no cumplirse la condición aterior entra este elif que es un else y if comprimido y esta condición se cumplira mientras k sea mayor igua a a y k menor a b
        if abs(l-t)<=2:          #Entra otro condicional que pide que si el valor absoluto de la resta de l y t es menor igual a 2 entonces haga las acciones de abajo:
            l+=1.2               #Al cumplirse la condicion sumamos a la variable l 1.2
            t+=0.6               #Al cumplirse la condicion sumamos a la variable t 0.6
        else:                    #Si no se cumple la condición anterior entra esta que es su contraria de if
            l+=0.2               #Al cumplirse la condicion sumamos a la variable l 0.2
            t+=0.6               #Al cumplirse la condicion sumamos a la variable t 0.6
    elif k>=b and k<c:           #Salimos de las condicionl elif anterior  que al ya no cumplirse, entramos a otro elif que se cumplira mientras k sea mayor igual a b y k sea menor a c  
        t+=0.4                   #Al cumplirse la condicion sumamos a la variable l 0.4
        l+=0                     #Al cumplirse la condicion sumamos a la variable t 0
    elif k>=c and k<d:           #Al ya no cumplirse la condicion aterior entramos a este elif que pide que k sea mayor igual a c pero menor a d
        l+=1.65                  #Al cumplirse la condicion sumamos a la variable l 1.65
        t+=0.6                   #Al cumplirse la condicion sumamos a la variable t 0.6
    if l>=400 or t>=400:         #Usamos otro condicional que si l o t es menor igual a 400 realizara la accion de abajo
        break                    #Al cumplirse una o otra condición del if romperá el bucle
    k+=1                         #Nuestro contador de tiempo que en cada suceso a k se le sumara 1
print("La liebre recorrio", l, "metros, y la tortuga recorrio", t, "metros")  #Imprimira un mensaje con el recorrido de la liebre y la tortuga qu estan dadas por la variable l y t
if l>t:                          #Declaramos otro condicional que pide que si l es menor a t, es decir el recorrido de la liebre es mayor a la tortuga realizara la acción de abajo
      print("Gano la libre")     #Si se cumple la acción anterior imprimira que ganó la libre
else:                            #Si se cumple la condición contraria del if pasado es decir que t es mayor a l entonces hara la acción de abajo
      print("Gano la tortuga")   #Si se cumple la acción anterior imprimira que ganó la tortuga
print("La carrera duró", k , "segundos")    #Imprimira un mensaje con el valor de nuestro contador de tiempo
    
```

El origen de las palabras.

En una biblioteca se han registrado las palabras que son mas comunes y unicas en los idiomas ingles y español. Dentro de los ejemplares del acervo se encuentra una bitacora desarrollada entre los años 1920 y 1960, esta bitacora cuenta con elementos de ambos idiomas, afortunadamente se tiene certeza de las palabras usadas por el autor en cada año.

La biblioteca cataloga a sus ejemplares por idioma, tu trabajo sera contar cuantas palabras hay de los dos idiomas en la bitacora; el ejemplar se clasificara en en idioma con mas elementos. Ademas para formalizar tu trabajo, has dos nuevas listas donde especifiques por cada año cuales palabras fueron utilizadas del ingles  y cuales del español.

Para facilitar el trabajo, se te proporciona la siguiente descripcion:


- La lista "bitacora"  tiene 40 "sublistas", cada una corresponde a un año de elaboracion.

- El primer elemento de cada sublista es el año de elaboracion, los demas elementos son las palabras que se usaron en ese mismo año.  La cantidad de elementos en cada sublista es variable.

- La biblioteca te proporciona la lista de las palabras mas comunes en ingles y en español. Para saber a cual idioma pertenece cada palabra de la bitacora, basta ver si esta palabra esta en la lista del ingles o en la lista del español. 


$\textbf{Sugerencia:}$ Crea dos contadores, uno contara las palabras del ingles  y otro las de español.  Crea dos bucles\clicos, uno se movera en un intervalo de años,  mientras que la variable del segundo ciclo/bucle se movera en los elementos de cada sublista.

Corre los siguientes codigos para que conozcas la lista de palabras de los dos idiomas y la composicion de la bitacora.

```python
#### Las palabras mas comunes de cada idioma son las siguientes:

palabras_ingles = ['the', 'of', 'and', 'to', 'in', 'a', 'that', 'is', 'it', 'for',
       'was', 'i', 'as', 'he', 'with', 'not', 'his', 'be', 'on', 'by',
       'this', 'you', 'at', 'or', 'from', 'are', 'but', 'have',
       'which', 'had', 'they', 'we', 'an', 'her', 'all', 'their',
       'one', 'were', 'she', 'my', 'so', 'him', 'if', 'when', 'there',
       'can', 'who', 'will', 'would', 'what', 'no', 'been', 'more',
       'them', 'has', 'do', 'me', 'its', 'other', 'out', 'into', 'time',
       'may', 'our', 'your', 'up', 'some', 'these', 'about', 'than',
       'said', 'then', 'only', 'could', 'such', 'any', 'two', 'like',
       'see', 'new', 'first', 'did', 'also', 'now', 'after', 'how',
       'most', 'man', 'should', 'very', 'us', 'over', 'life', 'made',
       'many', 'god', 'those', 'well', 'where', 'before']


palabras_espanol = ['de', 'la', 'en', 'y', 'el', 'que', 'los', 'del', 'se', 'las',
       'por', 'un', 'con', 'una', 'no', 'para', 'su', 'es', 'al', 'como',
       'lo', 'más', 'sus', 'pero', 'sobre', 'este', 'si', 'entre',
       'le', 'sin', 'esta', 'son', 'ha', 'ser', 'cuando', 'suyo', 'también',
       'todo', 'ya', 'era', 'puede', 'había', 'dos', 'desde', 'fue',
       'parte', 'así', 'muy', 'años', 'hasta', 'mi', 'todos', 'nos',
       'está', 'vida', 'mismo', 'estado', 'cada', 'vez', 'porque',
       'tiempo', 'otros', 'sólo', 'forma', 'han', 'qué', 'tanto', 'bien',
       'hay', 'tiene', 'caso', 'ni', 'donde', 'uno', 'otro',
       'general', 'él', 'sino', 'trabajo', 'ese', 'estos', 'social',
       'durante', 'después', 'otra', 'aunque', 'mayor', 'yo', 'ley',
       'lugar', 'siempre', 'mundo', 'gran', 'esto', 'sido', 'menos',
       'esa', 'poder', 'derecho', 'política']

#una lista con todas las palabras del inges y todas las del español
juntas = palabras_espanol + palabras_ingles


years    = arange(1920, 1961,1)
longitud = arange(30,  101, 1)

bitacora = []
for i in range(0,len(years)):
    semilla  = (4*i)**2
    seed(int(semilla))
    tamano   = choice(longitud)
    sublista = sample(juntas, k = tamano)
    sublista.insert(0,years[i])
    bitacora.insert(i,sublista)
```

```python
ingles,espanol=0,0            #Declaramos 2 variables iguales a 0
for i in range(len(bitacora)):   #Creamos un bucle for que que recorrera a la variable i en el rango de la longitud de la lista bitacora
    sublst=bitacora[i]           #Creamos una variable igual al elemento i de bitacora, es decir una sublista
    en,es=0,0                    #Creamos las variables en,es iguales a 0
    for j in range(1,len(sublst)):   #Creamos otro bucle for que recorrera la variable j en el rango de 1 y la longitud de la sublista
        palabra=sublst[j]          #Creamos una variable igual al elemento j de la sublista, es decir un elemnto de la sublista
        if palabra in palabras_ingles:   #Condicional que pide que si la palabra este en la lista palabras_ingles hará las acciones de abajo
            ingles+=1               #Al cumplirse la condicion se sumamos 1 a la variable ingles
            en+=1                   #Al cumplirse la condicion se sumamos 1 a la variable en
        elif palabra in palabras_espanol:   #Al ya no cumplirse la condición de abajo entrá este elif que pide que si la variable palabra esta en palabras_espanol hará las acciones de abajo
            espanol+=1           #Al cumplirse la condicion se sumamos 1 a la variable español
            es+=1                #Al cumplirse la condicion se sumamos 1 a la variable es
    if es<en:                 #Salimos de un for, pero seguimoes en el otro, y entra un if que pide que si es es mayor que en entonces haga la accion de abajo
        print("Para el año", sublst[0], "se registraron más palabras en inglés") #Al cumplirse la condicion del if se imprimira un mensaje con el elemento 0 de la sublista
    else:        #Si no se cumple el if se cumple la contraria, entonces entra este else, que realiza lo de abajo
        print("Para el año", sublst[0], "se registraron más palabras en español") #Al cumplirse la condicion del else se imprimira un mensaje con el elemento 0 de la sublista
if espanol<ingles:   #otra condicional que pide que si la variable espanos es menor a ingles se haga la condición de abajo
    print("La bitacora està escrita en inglés")   #Al cumplirse la condicion del if se imprimira un mensaje
else:           #Si no se cumple el if se cumple el contrario y entra este else
    print("La bitacora està escrita en español") #Al cumplirse la condicion del else se imprimira un mensaje
```
